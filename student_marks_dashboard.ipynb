```python
import matplotlib.pyplot as plt
Students= ["Muskan","Shweta","Trisha","Riya","Shreya"]
Marks= [90,95,90,80,70]
plt.bar(Students,Marks, color= "blue")
plt.title("Students marks dasboard")
plt.xlabel("Stdents")
plt.ylabel("Marks")
plt.show()
```

```python
plt.hist(Marks, bins=5, color= "purple")
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Nmber of students")
plt.show()
```
