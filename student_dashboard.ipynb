```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data= {
    "Name": ["Muskan", "anjali", "riya", "shweta", "priya", "neha"],
    "Course": ["Python", "Java", "DBMS", "Python", "C++", "Java"],
    "Marks": [85, 90, 78, 92, 88, 95]
}
df= pd.DataFrame(data)
print(df)
sns.set_theme(style="darkgrid")
sns.countplot(x= "Course", data= df)
plt.title("Count of Students by Course")
plt.xlabel("Course")
plt.ylabel("Count")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.boxplot(y= "Marks", data= df)
plt.title("Marks Distribution")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.histplot(x= "Marks", data= df, bins= 7)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```
