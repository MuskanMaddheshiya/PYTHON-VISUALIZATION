```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sales_dataset.csv")
print(df)

df["Revenue"] = df["Sales"]
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.month_name()

# ---------------- Product-wise Sales Bar Chart ----------------
product_sales = df.groupby("Product Name")["Sales"].sum()

plt.figure(figsize=(8,5))
product_sales.plot(kind="bar", color="skyblue")
plt.title("Product-wise Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")
plt.show()
```

```python

# ---------------- Monthly Revenue Line Chart ----------------
monthly_revenue = df.groupby("Month")["Revenue"].sum()

plt.figure(figsize=(8,5))
monthly_revenue.plot(kind="line", marker="o", color="green")
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
# ---------------- Sales Histogram ----------------
plt.figure(figsize=(8,5))
plt.hist(df["Sales"], bins=5, edgecolor="black", color="orange")
plt.title("Sales Histogram")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# ---------------- Sales vs Profit Scatter Plot ----------------
plt.figure(figsize=(8,5))
plt.scatter(df["Sales"], df["Profit"], color="red")
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
