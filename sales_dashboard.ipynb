```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data= {
    "Months": ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"],
    "Sales": [50000,80000,45000,90000,65000,85000,70000,83000,58000,95000,60000,77000],
    "Profit": [10000,15000,8000,20000,12000,18000,14000,16000,11000,22000,13000,17000]
}
df= pd.DataFrame(data)
print(df)
sns.set_theme(style="darkgrid")
plt.figure(figsize=(12,12))
sns.lineplot(x= "Months", y= "Sales", data= df, color= "blue", marker= "o")
plt.title("Monthly Sales dashboard")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```
