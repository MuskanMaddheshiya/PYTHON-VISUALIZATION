```python
import matplotlib.pyplot as plt
Months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
Sales = [50000,80000,45000,90000,65000,85000,70000,83000,58000,95000,60000,77000]
plt.plot(Months,Sales, color= "green", linestyle="--", marker= "o")
plt.title("Monthly sales report")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
plt.bar(Months,Sales, color= "purple")
plt.title("Monthly sales record")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(Sales,labels= Months,autopct='%1.1f%%')
plt.title("Monthly sales")
plt.show()
```
