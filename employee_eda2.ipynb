```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("employee-salary.xlsx")

# Histogram
plt.figure(figsize=(8,5))
plt.hist(df["Monthly_Salary"], bins=5, color="purple", edgecolor="black")
plt.title("Salary Distribution Histogram")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.savefig("histogram.png")
plt.show()

# Count Plot
plt.figure(figsize=(8,5))
sns.countplot(x="Department", data=df, palette="Set2")
plt.title("Department Count Plot")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.savefig("countplot.png")
plt.show()

# Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(range(len(df)), df["Monthly_Salary"], color="green")
plt.title("Employee Index vs Monthly Salary Scatter Plot")
plt.xlabel("Employee Number")
plt.ylabel("Monthly Salary")
plt.savefig("scatterplot.png")
plt.show()

# Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(y="Monthly_Salary", data=df, color="green")
plt.title("Monthly Salary Box Plot")
plt.savefig("boxplot.png")
plt.show()

# Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[["Monthly_Salary"]].corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap")
plt.savefig("heatmap.png")
plt.show()
```
